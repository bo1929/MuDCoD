```python
import yaml
import numpy as np
import pandas as pd
import seaborn as sns
import tikzplotlib
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from mudcod.nw import Similarity  # noqa: E402
```

```python
cell_name = "DA"

adj_path = f"../data/cell-type-specific/{cell_name}-networks/adj/p95/multisubject_dynamic_nw_series.npy"
adj_ms_series = np.load(adj_path)
```

```python
membership_distances = np.zeros((adj_ms_series.shape[0], adj_ms_series.shape[1], adj_ms_series.shape[1]))
for sbj in range(membership_distances.shape[0]):
    for ti in range(adj_ms_series.shape[1]):
        for tj in range(ti+1, adj_ms_series.shape[1]):
            membership_distances = Similarity.ipsen_mikhailov_distance(
                    adj_ms_series[sbj, ti, :,:], adj_ms_series[sbj, tj, :,:]
                )
            print(f"sbj:{sbj} btwn {ti}-{tj} :: {membership_distances}")
    
```
